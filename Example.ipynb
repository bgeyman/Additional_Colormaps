_____________
Make a list of hex codes using `make_hex_list()` and then create custom 256 color **LinearSegmentedColormap** object using `make_custom_cmap_from_hex_list()`

```python
rgb_list = [[175,162,116], [30,72,105]]
test_hex_list = make_hex_list(rgb_list)
make_custom_cmap_from_hex_list(hex_list=test_hex_list, n_colors=256, cmap_name='test')
```

The above functions can be also be used to interpolate limited color schemes as below:

```python
rgb_values = (np.array(new_cmaps['perc2_9lev'].colors)[:,:3]*255).astype(int)
test_hex_list = make_hex_list(rgb_values)
make_custom_cmap_from_hex_list(hex_list=test_hex_list, n_colors=256, cmap_name='interpolated_perc2_256lev')
```
